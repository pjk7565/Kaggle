# 2. Natural Language Processing

In almost all NLP, one will generally always have to undergo these few pre-processing steps to convert the input raw text into a form that is readable by your model and the machine. Behind the scenes, text pre-processing can be boiled down to these few simple steps:  
  
1. **<span style='color:#20BEFF'>Tokenization</span>**: Segregation of the text into its individual constitutent words.
2. **<span style='color:#20BEFF'>Stopwords</span>**: Throw away any words that occur too frequently as its frequency of occurrence will not be useful in helping detecting relevant texts. (as an aside also consider throwing away words that occur very infrequently)
3. **<span style='color:#20BEFF'>Stemming</span>**: Combine variants of words into a single parent word that still conveys the same meaning.
4. **<span style='color:#20BEFF'>Vectorization</span>**: Convering text into vector format. One of the simplest is the famouse bag-of-words approach, where you create a matrix (for each document or text in the corpus). In the simplest form, this matrix stores word frequencies (word counts) and is often referred to as vectorization of the raw text.  
  
**<span style='color:#20BEFF'>Natural Language Toolkit (NLTK)</span>**: To make our NLP endeavours more convenient, let me introduce to you one of the most handy toolkits that on NLP.

```python
import nltk
nltk.download('punkt')
```

## 2.1 Tokenization

The concept of tokenization is the act of taking a sequence of characters in a given document and dicing it up into its individual constituent pieces, which are the eponymous 'tokens' of this method. One could loosely think of them as singular words in a sentence. One could naively implement the 'split()' method on a string which separates it into a python list based on the identifier in the argument. It is actually not that trivial.  
  
Here we split the first sentence of the text in the training data just on a space as follows:

```python
first_text = train.text.values[0]
print(first_text)
print('=' * 90)
print(first_text.split(' '))
```

However as you can see from this first attempt at tokenization, the segregation of the sentence into its individual elements (or terms) is not entirely accurate. (example: 'process,') The punctuation mark(comma) has also been included and is being treate along with the word 'process' as a term in itself. Ideally we would like the comma and the word to be in two different and separate elements of the list. Trying to do this with pure python list operations will be quite complex so this is where the NLTK library comes into play. There is a convenient method 'word_tokenize()' which strips out singular words as well as punctuations into separate elements automatically as follows:

```python
first_text_list = nltk.word_tokenize(first_text)
print(first_text_list)
```

## 2.3 Stemming and Lemmatization

Stemming attempts to reduce as many different variations of similar words into a single term (different branches all reduced to single word stem). Therefore if we have 'running', 'runs' and 'run', you would really want these three distinct words to collapse into just the word 'run'.  
  
We can turn to NLTK again which provides various stemmers which include variants such as the Porter stemming algorithm, the lancaster stemmer and the Snowvall stemmer. In the following example, I will create a porter stemmer instance as follows:

```python
stemmer = nltk.stem.PorterStemmer()
```

```python
# example

print('The stemmed form of running is: {}'.format(stemmer.stem('running')))
print('The stemmed form of runs is: {}'.format(stemmer.stem('runs')))
print('The stemmed form of run is: {}'.format(stemmer.stem('run')))
```

As we can see, the stemmer has successfully reduced the given words above into a base form and this will be most in helping us reduce the size of our dataset of words when we come to learning and classification tasks.  
  
However there is one flaw with stemming and that is the fact that the processing involves quite a crude heuristic in chopping off the ends of words in the hope of reducing a particular word into a human recongizable base form. Therefore this process does not take into account vocabulary or word forms when collapsing words as this example will illustrate:

```python
print('The stemmed form of leaves is: {}'.format(stemmer.stem('leaves')))
```

### 2.3.1 Lemmatization to the rescue

Therefore we turn to another that we could use in lieu of stemming. This method is called lemmatization which aims to achieve the same effect as the former method. However unlike a stemmer, lemmatizaing the dataset aims to reduce words based on an actual dictionary or vocabulary (the Lemma) and therefore will not chop off words into steemed forms that do not carry any lexical meaning. Here we can utilize NLTK once again to initialize a lemmatizer (WordNet variant) and inspect how it collapses words as follows:

```python
nltk.download('wordnet')
```

## 2.4 Vectorizing Raw Text

We must now think of how to feed the raw text into a machine learning model. Having already discussed tokenization, stopword removals and stemming (or lemmatizing) we have now arrived at a reasonably human readable is still unfortunately not yet machine readable. A machine can read in bits and numbers and therefore we will first need to convert our text into numbers for which we utilise a very common approach known as the Bag-of-Words.

**<span style='color:#20BEFF'>The Bag of Words approach</span>**  
  
This approach uses the counts of words as a starting block and records the occurrence of each word (form the entire text) in a vector specific to that particular word. For example given these two sentences 'I love to eat Burgers', 'I love to eat Fries', we first tokenize to obtain our vocabulary of 6 words from which we can get the word counts for - [I, love, to, eat, Burgers, Fries]  
  
Vectorizing the text via the Bag of Words approach, we get six distinct vectors one for each words. So you ask since we now have rows consisting of numbers (instead of text) what forms the columns (or features)? Each word now becomes an individual feature/column in this new transformed dataset. To illustrate this point, I shall utilize the Scikit-learn library to implemnet a vectorizer that generates a vector of word counts (term frequencies) - via the CountVectorizer method as follows:

```python
sentence = ['I love to eat Burgers', 'I love to eat Fries']
vectorizer = CountVectorizer(min_df=0)
sentence_transform = vectorizer.fit_transform(sentence)
```

**<span style='color:#20BEFF'>Fitting the vectorizer to the dataset</span>**  
  
The parameters that I have provided explicitly (the rest are left as dafault) are the bare minimum. Here 'min_df' in the parameter refers to the minimum document frequency and the vectorizer will simply drop all words that occur less than that value set (either integer or in fraction form). Finally we apply the fit_transform method is actually comprised of two steps. The first step is the fit method where the vectorizer is mapped to the dataset that you provide. Once this is done, the actual vectorizing operation is performed via the transform method where the raw text is turned into its vector form as shown below:

# 3. Topic modeling

I will implement two different topic modeling techniques as follows:  
  
1. **<span style='color:#20BEFF'>Latent Dirichlet Allocation</span>**: Probabilistic, generative model which uncovers the topics latent to a dataset by assigning weights to words in a corpus, where each topic will assign different probaility weights to each word.
2. **<span style='color:#20BEFF'>Non-negative Matrix Factorization</span>**: Approximation method that takes an input matrix and approximates the factorization of this matrix into two other matrices, with the caveat that the values in the matrix be non-negative.

```python
# define helper function to print top words
def print_top_words(model, feature_names, n_top_words):
    for index, topic in enumerate(model.components_):
        message = '\nTopic #{}: '.format(index)
        message += ' '.join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
        print('='* 70)
```

## 3.1 Putting all the preprocessing steps together

This is now the perfect opportunity to piece together all the text preprocessing steps that was mentioned in the previous section. However, we do not need to go through all the steps in defining tokenization, stopword removals, stemming/lemmatizing.  
  
When you vectorize the raw text with CountVectorizer, the dual stages of tokenizing and stopwords filtering are automatically included as a high-level component. Here unlike the NLTK tokenizer that you were introduced to in the Section 2.1 earlier, Sklearn's tokenizer discards all single character terms like('a', 'w' etc) and also lower cases all terms by default. Filtering out stopwords in Sklearn is as convenient as passing the value 'english' into the argument 'stop_words' where a built-in English stopword list is automatically used.  
  
Unfortunately, there is no built-in lemmatizer in the vectorizer so we are left with a couple of options. Either implementing it separately everytime before feeding the data for vectorizing or somehow extend the sklearn implementation to include this functionality. Lucily for us, we have the latter option where we can extent the CountVectorizer class by overwriting the 'build_analyzer' method as follows:

### 3.1.1 Extending the CountVectorizer class with a lemmatizer

```python
lemm = WordNetLemmatizer()

class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc: (lemm.lemmatize(w) for w in analyzer(doc))
```

We have essentially inherited and subcalssed the original Sklearn's CountVectorizer class and overwritten build_analyzer method by implementing the lemmatizer for each list in the raw text matrix.

```python
# storing the entire training text in a list
text = list(train.text.values)

# calling our overwritten Count vectorizer
tf_vectorizer = LemmaCountVectorizer(
    max_df=0.95, min_df=2, stop_words='english', decode_error='ignore'
)
tf = tf_vectorizer.fit_transform(text)
```

## 3.2 Latent Dirichlet Allocation

Finally we arrive on the subject of topic modeling and the implementation of a couple of unsuperviesed learning algorithms. The first method that I will touch upon is Latent Dirichlet Allocation. Now there are a couple of different implements of this LDA algorithm but in this notebook, I will be using Sklearn's implementation. Another very well-known LDA implementation is Radim Rehurek's gensim, so check it out as well.

**<span style='color:#20BEFF'>Corpus - Document - Word: Topic Generation</span>**  
  
In LDA, the modeling process revolves around three things: the text corpus, its collection of documents, D and the words W in the documents. Therefore the algorithm attempts to uncover K topics from this corpus via the following way.  
  
The LDA algorithm first models documents via a mixture model of topics. From these topics, words are then assigned weights based on the probability distribution of these topics. It is this probabilitic assignment over words that allow a user of LDA to say how likely a particular word falls into a topic. Subsequently from the collection of words assigned to a particular topic, are we thus able to gain an insight as to what that topic may actually represent from a lexical point of view.  
  
parameters:  
1. n_components: The number of topics that you specify to the model
2. $\alpha$ parameter: This is the dirichlet parameter that can be linked to the document topic prior
3. $\beta$ parameter: This is the dirichlet parameter linked to the topic word prior  
  
To invoke the algorithm, we simply create an LDA instance through the Sklearn's LatentDirichletAllocation function. The various parameters would ideally have been obtained through some sort of vaildation scheme. In this instance, the optimal value of n_components (or topic number) was found by conducting a KMeans + Latent Semantic Analysis Scheme (as shown in this paper here) whereby the number of KMeans clusters and number of LSA dimensions were iterated through and the best silhouette mean score.

```python
lda = LatentDirichletAllocation(
    n_components=11, max_iter=5, learning_method='online',
    learning_offset=50., random_state=0
)
```

```python
lda.fit(tf)
```

### 3.2.1 Topics generated by LDA
